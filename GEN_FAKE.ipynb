```python
from gen_models import AssortmentGenerator, MC
import numpy as np
import random
import os


Vec_Len = 21


PATH = "BCG_fake"
rho_PATH = "BCG_rho_lamb/rhoBCG_NN.npy"
lamb_PATH = "BCG_rho_lamb/lambBCG_NN.npy"



seed = 5678
np.random.seed(seed)
random.seed(seed)

N_sample = 200000




if not os.path.exists(PATH):
    os.makedirs(PATH)

dir = os.listdir(PATH)
if len(dir) > 0:
    print("Non empty directory")
    print("remove files!")
    for f in dir:
        os.remove(os.path.join(PATH, f))


AssortmentGen = AssortmentGenerator.GenAssortment["Abundant"]


rho = np.load(rho_PATH)
lamb = np.load(lamb_PATH)

   
TrueModel = MC(Vec_Len)
TrueModel.M = rho
TrueModel.Lam = lamb


TrueModel.save_para(PATH)


INPUT = np.zeros((N_sample,Vec_Len))
SAMP_OUTPUT = np.zeros((N_sample,Vec_Len))

for i in range(N_sample):

    assort = AssortmentGen(Vec_Len)
    assort[0] = 1


    INPUT[i] = assort

    SAMP_OUTPUT[i] = TrueModel.gen_final_choice(assort)




np.save(PATH+"/"+"ASSORT",INPUT)
np.save(PATH+"/"+"SAMP",SAMP_OUTPUT)



```
